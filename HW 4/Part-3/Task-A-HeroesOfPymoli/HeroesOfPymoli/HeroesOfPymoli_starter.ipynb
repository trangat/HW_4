```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = purchase_data["Item ID"].nunique()
#print(unique_items)

average_price = purchase_data["Price"].mean()
#print(average_price)

purchases = purchase_data["Purchase ID"].count()
#print(purchases)

total_rev = purchases*average_price
#print(total_rev)

#create summary df with all above values 

summary_purchases = pd.DataFrame({"Number of Unique Items": [unique_items], 
              "Average Price": average_price, 
              "Number of Purchases": purchases, 
              "Total Revenue": total_rev})

summary_purchases["Average Price"] = summary_purchases["Average Price"].map("${:,.2f}".format)
summary_purchases["Total Revenue"] = summary_purchases["Total Revenue"].map("${:,.2f}".format)

summary_purchases
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed


```python
#Use groupby to sort through initial df 
group_df = purchase_data.groupby(["Gender"])

grouped_gender = group_df.nunique()

#calculate total gender, then male gender, then female gender, and other/non-disclosed
total_gender = grouped_gender["SN"].sum()
#total_gender
gender_count = grouped_gender["SN"].unique()
#gender_count

#calculate percentages & format df

percentage_gender = (grouped_gender["SN"])/total_gender
#percentage_gender

gender_analysis_df = pd.DataFrame({"Total Count": gender_count,
                                   "Percentage of Players": percentage_gender})

gender_analysis_df["Percentage of Players"] = gender_analysis_df["Percentage of Players"].map("{:,.2%}".format) 

gender_analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#drop duplicates first
clean_data = purchase_data.drop_duplicates("SN")
clean_data.head()

# Create the bins in which Data will be held  
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

# Create the names for the  bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#create "age groups" column and groupby column 
clean_data["Age Groups"] = pd.cut(clean_data["Age"], bins, labels=group_names)
clean_data.head()

#calculate # of people in each age group 
ages = clean_data["Age Groups"].value_counts()
ages

#calculate percentage of players in each age group 
player_percent = ages/total_players
player_percent

#create df and format 
age_demo_df = pd.DataFrame({"Total Count": ages,
                            "Percentage of Players": player_percent})
age_demo_df["Percentage of Players"] = age_demo_df["Percentage of Players"].map("{:,.2%}".format)

#order bins 
age_demo_df.sort_index()

```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
total_age_number = age_demo_df["Total Count"]

#Create the bins in which Data will be held  
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

# Create the names for the  bins
groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#create "age groups" column and group by column 
purchase_data["Age Groups"] = pd.cut(purchase_data["Age"], bins, labels=groups)
purchase_data.head()

#calculate total purchases, avg. purchase price, avg. purchase total/person by age

total_age_purchases = purchase_data.groupby(["Age Groups"]).count()["Purchase ID"]
total_age_purchases 
                                        
age_price = purchase_data.groupby(["Age Groups"]).sum()["Price"]
age_price

#total_value = total_purchases*age_price
#total_value
#figure out avg. total/person

avg_total = age_price/total_age_purchases
avg_total

avg_purchase = age_price/total_age_number
avg_purchase

purchase_analysis_df = pd.DataFrame({"Purchase Count": total_age_purchases,
                                     "Average Purchase Price": avg_total,
                                     "Total Purchase Value": age_price, 
                                     "Average Purchase/Person": avg_purchase})

purchase_analysis_df["Average Purchase Price"] = purchase_analysis_df["Average Purchase Price"].map("${:,.2f}".format)
purchase_analysis_df["Total Purchase Value"] = purchase_analysis_df["Total Purchase Value"].map("${:,.2f}".format)
purchase_analysis_df["Average Purchase/Person"] = purchase_analysis_df["Average Purchase/Person"].map("${:,.2f}".format)

purchase_analysis_df

purchase_analysis_df.sort_index()
```
